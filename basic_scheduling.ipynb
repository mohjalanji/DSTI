```python
import matplotlib.pyplot as plt
import numpy as np

# Number of tasks and cores
num_tasks = 10
num_cores = 4

# Generating random intervals for tasks
np.random.seed(0)
tasks = np.random.randint(10, 100, size=(num_tasks, 2))
tasks = np.sort(tasks, axis=1)
tasks[:, 1] += tasks[:, 0]  # End time

# Colors for tasks
colors = plt.cm.viridis(np.linspace(0, 1, num_tasks))

# Setting up the plot
fig, ax = plt.subplots(figsize=(10, 6))

for i in range(num_tasks):
    start, end = tasks[i]
    core = i % num_cores + 1
    ax.plot([start, end], [core, core], color=colors[i], linewidth=4)
    ax.plot([start, start], [core - 0.1, core + 0.1], color='black')

# Setting axis labels and limits
ax.set_xlabel('Time')
ax.set_ylabel('Core')
ax.set_xticks(np.linspace(0.5, 1000, 10))
ax.set_xlim(0.5, 1000)
ax.set_yticks(np.arange(1, num_cores + 1))
ax.set_ylim(0.5, num_cores + 0.5)
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

plt.title('Schedule Table with Linked Task Intervals')
plt.show()
```
